```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('spotify_songs.csv')
```

```python
# dropping duplicates
df = df.drop_duplicates('track_name', keep='first')

# sort by top 100 tracks
df = df.sort_values('track_popularity', ascending=False).head(100)

# drop all non used cols
cols_to_drop = ['track_id', 'track_popularity', 'track_album_id', 'playlist_name','playlist_id','danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms']
df = df.drop(columns=cols_to_drop)
```

```python
# make half of playlist subgenre null for optional field
for index, row in df.iterrows():
    if np.random.rand() < 0.5:
        df.loc[index, 'playlist_subgenre'] = np.nan
```

```python
# resetting index, might need to change 
df.set_index('track_name', inplace=True)
```

```python
# create json
df.to_json('spotify_100_songs.json')
```
